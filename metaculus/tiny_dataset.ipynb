```python
import json

with open('../data/forecast_data/question_set.json') as f:
    question_data = json.load(f)

with open('../data/forecast_data/resolution_set.json') as f:
    resolution_data = json.load(f)

resolution_data = [res for res in resolution_data['resolutions'] if res['resolved']]

#filter for questions not containing key phrases
wordbank = ["close date", "yield", "index", "stock", "price", "market", "trading", "exchange", "currency", "dollar", "euro", "yen", "pound", "rupee", "bitcoin", "ethereum"]
print(question_data['questions'][0])

# print(len(question_data))
# resolution_data = resolution_data[::9][:20]  # Limit 20 resolved questions for evaluation
res_id = {res['id'] for res in resolution_data}
question_data = [q for q in question_data['questions'] if q['id'] in res_id]
question_data = [q for q in question_data if not any(word in q['question'].lower() for word in wordbank)]

joint_data = []
res_dict = {res['id']: res for res in resolution_data}
for q in question_data:
    qid = q['id']
    res = res_dict[qid]
    q_copy = q.copy()
    q_copy['resolution_date'] = res['resolution_date']
    q_copy['resolved_to'] = res['resolved_to']
    joint_data.append(q_copy)

joint_data = joint_data[::5]  # Limit to 20 questions for evaluation

with open('data/tiny_dataset.json', 'w') as f:
    json.dump(joint_data, f, indent=2)
```
