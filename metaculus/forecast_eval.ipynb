```python
import json

with open('../data/forecast_data/question_set.json') as f:
    question_data = json.load(f)

with open('../data/forecast_data/resolution_set.json') as f:
    resolution_data = json.load(f)
```

```python
question_ids = [question['id'] for question in question_data['questions']]
resolution_qid = [res['id'] for res in resolution_data['resolutions']]

print("Question IDs:", len(question_ids))
print("Resolution IDs:", len(resolution_qid))
print("Intersection:", len(set(question_ids) & set(resolution_qid)))

```
